### Figure 3B

For this figure I am going to use only a few compatible results with the model. This values will be added to a multi-index dataframe taking the values from Figure 2B. This dataframe is named ``exp_data`` from *experimental data* and it will be used to parametric adjustment for the extended Nieto *et al.* (2022) model.

```python
extra_cols = pd.DataFrame()
extra_cols['Light'] = (['Low light'] * 4 + ['Dark'] * 4)
extra_cols['Treatment'] = (['Mock', '10 muM ABA'] * 4)

# Concatenar horizontalmente
exp_data = pd.concat([extra_cols, df1_trimmed.copy().reset_index(drop=False)], axis=1)

# Crear un MultiIndex
exp_data.set_index(['Light', 'Treatment'], inplace=True)

# Añadir la columna 'index' como tercer nivel del índice
exp_data.set_index('index', append=True, inplace=True)

# Renombrar los niveles del índice
exp_data.index.set_names(['Light_condition', 'ABA_treatment', 'Metric'], inplace=True)



# Ordenar dataframe
from itertools import product
# Definir los niveles del índice
lights = ['Low light', 'Dark']; treatments = ['Mock', '10 muM ABA']; metrics = ['Average', 'SD']
desired_order = list(product(lights, treatments, metrics))
exp_data = exp_data.reindex(desired_order)


# Añadir datos de Fig. 3B:
#                     LightMockAv, LightMockSD, LightABAAv,  LightABASD,  Dark...
exp_data['phyB-9'] =    5.752222,    0.531432,    2.202222,    0.430255,    np.nan, np.nan, np.nan, np.nan
exp_data['35S:PIF4'] =  13.129625,   1.273713125, 7.151333333, 1.151332272, np.nan, np.nan, np.nan, np.nan

exp_data
```

```python
# Guardar
exp_data.to_csv('experimental_data_for_extended_model.csv', index=True)
exp_data
```
